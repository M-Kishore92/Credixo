# 01. Exploratory Data Analysis (EDA)

This notebook explores the synthetic dataset, analyzing distributions, correlations, and missing values.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../backend/data/synthetic_loan_data.csv')
print(df.head())
print(df.info())

# Missing Values
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Value Heatmap')
plt.show()

# Class Balance
df['loan_repaid'].value_counts().plot(kind='bar')
plt.title('Loan Repaid Class Balance')
plt.show()
```
