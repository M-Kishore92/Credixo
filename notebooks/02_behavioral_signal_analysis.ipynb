# 02. Behavioral Signal Analysis

Validating the predictive power of alternative behavioral signals against loan repayment.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../backend/data/synthetic_loan_data.csv')

# Boxplots
sns.boxplot(x='loan_repaid', y='electricity_payment_regularity', data=df)
plt.title('Elec Regularity vs Repayment')
plt.show()

sns.boxplot(x='loan_repaid', y='utility_payment_consistency', data=df)
plt.title('Utility Consistency vs Repayment')
plt.show()
```
