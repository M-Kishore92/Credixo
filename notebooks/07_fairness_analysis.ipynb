# 07. Fairness Analysis

Measuring demographic parity and equal opportunity across gender and area type groups.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../backend/data/synthetic_loan_data.csv')

# Approval rates by group (simulated)
summary = df.groupby(['gender', 'area_type'])['loan_repaid'].mean().unstack()
summary.plot(kind='bar', figsize=(10, 6))
plt.title('Simulated Approval Rates by Gender and Area')
plt.ylabel('Approval Rate')
plt.show()
```
